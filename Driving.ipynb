```python
import pandas as pd

# Load driving log (CSV with image paths and steering angles)
df = pd.read_csv('driving_log.csv')

print(df.head())
```

```python
import os

print("Columns in dataset:", df.columns.tolist())
print("Steering angle statistics:")
print(df['steering'].describe())
```

```python
import numpy as np
from sklearn.metrics import mean_squared_error

mean_angle = df['steering'].mean()
y_true = df['steering']
y_pred = [mean_angle] * len(y_true)

mse = mean_squared_error(y_true, y_pred)
print("Mean Steering Angle:", mean_angle)
print("MSE (Baseline):", mse)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('/content/driving_log.csv')
plt.hist(df['steering'],bins=50)
plt.title("steering angle with distribution")
plt.show()
```

```python
import seaborn as sns

left_turns = df[df['steering'] < -0.15]
right_turns = df[df['steering'] > 0.15]

filtered_df = pd.concat([left_turns, right_turns])

plt.figure(figsize=(10, 6))
sns.histplot(filtered_df['steering'], bins=50, kde=True)
plt.title("Distribution of Steering Angles")
plt.xlabel("Steering Angle")
plt.ylabel("Frequency")
plt.show()
```
